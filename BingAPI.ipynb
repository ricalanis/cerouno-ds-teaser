```python
import random
from py_ms_cognitive import PyMsCognitiveImageSearch
import os
from requests import get
from slugify import slugify
from PIL import Image
```

```python
API_KEY = "[API-KEY]"
```

```python
make_model_dirs(["model_data/train", "model_data/test", "model_data/valid"])
download_bing_pictures("megaman x gameplay")
download_bing_pictures("super mario world gameplay")
```

```python
# Crear directorios para las fotografías
def make_model_dirs(folders):
    for folder in folders: make_dir(folder)
    return "OK"
```

```python
# Función para buscar las fotografías en el servicio web
def download_bing_pictures(keyword, quota = 300):
    search_service = PyMsCognitiveImageSearch(API_KEY, keyword)
    print("***Searching Pictures***")
    result_list = search_service.search_all(quota)
    # Agregar el resultado, filtrar solo jpeg y png
    result_list = [result.json for result in result_list if result.json["encodingFormat"] in ["jpeg", "png"]]
    keyword = slugify(keyword)
    print("***Saving Pictures***")
    make_model_dirs(["model_data/train/"+keyword, "model_data/test/"+keyword, "model_data/valid/"+keyword])
    for result in result_list: 
        try:
            save_file(keyword, destination_folder(), result["contentUrl"], result["encodingFormat"])
        except:
            print("***Fallo en descarga***")
    return "OK"
```

```python
# Función para decidir en qué carpeta va dicha fotografía
def save_file(keyword, directory,  url, ext):
    filename =  "model_data/"+directory+"/" + keyword + "/" + random_with_N_digits(10)+ "." + ext
    download(url, filename)
    filter_image(filename)
    print(filename)
    return "OK"
```

```python
def filter_image(filename):
    try:
        im = Image.open(filename)
        im.verify()
    except:
        os.remove(filename)
        print("****Error found, removing file****")
    return
```

```python
def destination_folder():
    r = random.uniform(0, 1)
    if r <= 0.6:
        return "train"
    elif r>0.6 and r<0.8:
        return "test"
    return "valid"
```

```python
def make_dir(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)
    return "OK"
```

```python
def random_with_N_digits(n):
    range_start = 10**(n-1)
    range_end = (10**n)-1
    return str(random.randint(range_start, range_end))
```

```python
def download(url, file_name):
    with open(file_name, "wb") as file:
        response = get(url)
        file.write(response.content)
```
